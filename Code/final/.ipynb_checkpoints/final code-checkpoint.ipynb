```python
import ctypes
import numpy as np
import copy
from scipy.signal import convolve2d

def sigmoid(x):
    #Vor.: -x- ist vom Typ np.darray (Numpy-Objekt für eine n-dimensionale Matrix)
    #Erg.: Die gesamte Matrix wird durch die Sigmoid-Funktio geführt. Das Ergebnis ist als Rückgabewert geliefert.
    
    return (1 / (1 + np.exp(-x)) )

def softmax(x):
    #Vor.: -x- ist vom Typ np.darray (Numpy-Objekt für eine n-dimensionale Matrix)
    #Erg.: Die gesamte Matrix wird durch die Softmax-Funktion geführt. Das Ergebnis ist als Rückgabewert geliefert.
    
    return np.exp(x) / sum(np.exp(x))

def RELU_Matrixoperation(inputMatrix):
    #Vor.: -x- ist vom Typ np.darray (Numpy-Objekt für eine n-dimensionale Matrix)
    #Erg.: Die gesamte Matrix wird durch die RELU-Funktion geführt. Das Ergebnis ist als Rückgabewert geliefert.
    
    return np.maximum(inputMatrix, 0)

def getinputStackedColumns(inputMatrix, kernel_map_shape, stride):
    #Vor.: -x- ist vom Typ np.darray (Numpy-Objekt für eine n-dimensionale Matrix)
    #Erg.: Für die Faltungsoperation o. das Pooling wird die Eingabematrix in Fenster eingeteilt, die sich ergeben, 
    #      wenn der Kernel über die Matrix gezogen wird. Dies wird benötigt, um im nächsten Schritt die Faltung bzw. das Pooling
    #      mit der Matrixoperation np.einsum() zu realisieren. Die verschiedenen Fenster fü die Eingabematrix
    #      -inputMatrix- auf Basis des Kernels mit den Dimensionen -kernel_map_shape- werden in Form eines np.array als Rückgabewert
    #      geliefert.
    #      Für mehr Informationen zu np.einsum(): https://numpy.org/doc/stable/reference/generated/numpy.einsum.html
    
    s0, s1, s2, s3 = inputMatrix.strides

    d1_input, d2_input, h_input, w_input = inputMatrix.shape
    h_kernel, w_kernel = kernel_map_shape

    out_shape = ( d1_input, d2_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride, h_kernel, w_kernel)
    inputStackedColumns = np.lib.stride_tricks.as_strided(inputMatrix,
                                                          shape=out_shape,
                                                          strides=(s0, s1, stride*s2,stride*s3,s2, s3))
    return inputStackedColumns

def getPadding(h_kernel, w_kernel, h_input, w_input, stride):
    #Vor.: Alle Eingabeparameter sind vom Typ int.
    #Erg.: Das Null-Padding (Rand mit 0) für die Breite und Höhe des Kernels, das
    #      benötigt wird, um für (h_input-h_kernel+1)//stride eine gerade Zahl herauszubekommen, ist
    #      ist als zweielementiger Tuple geliefert.
    
    initial_h = h_input
    initial_w = w_input 
    
    while True:
        if (h_input-h_kernel+1)%stride!=0:
            h_input+=1
            if h_input>=h_kernel:
                break
        else:
            break
            
    while True:
        w_input+=1
        if (w_input-w_kernel+1)%stride!=0:
            w_input+=1
            if w_input>=w_kernel:
                break
        else:
            break

            
    return h_input-initial_h, w_input-initial_w

def Convolution_strided_img2col(inputMatrix, kernel_map, stride):
    #Vor.: inputMatrix ist vom typ np.darray und hat 4 Dimensionen. kernel_map ist vom Typ np.darray und
    #      hat 3 Dimensionen.
    #Erg.: Dimension 3 der Eingabematrix entspricht allen Trainingsbeispielen. Dimension 4 entspricht allen Farbkanälen
    #      für die jeweiligen Trainingsbeispiele. Dimensionen 1 und 2 entsprechen den jeweiligen Pixelwerten der Farbkanäle.
    #      Für jedes Trainingsbeispiel wird jeder der 2d-Kernels in dem 3d-Kernel auf alle Farbkanäle angewandt (Faltung!). Alle 
    #      resultierenden Featuremaps sind als np.darray (4-dimensional) geliefert. Jeder Kernel liefert dabei eine neue Feature-Map.
    #      Die resultierenden Matrizen für jedes Trainingsbeispiel haben somit so viele (Farb)kanäle wie Kernels angewandt werden.
  
    d1_input, d2_input, h_input, w_input = inputMatrix.shape
    d_kernel, h_kernel, w_kernel = kernel_map.shape
    
    inputStackedColumns = getinputStackedColumns(inputMatrix, kernel_map.shape[1:], stride)
    
    out_shape = ( d1_input, d2_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride, h_kernel, w_kernel)

    inputStackedColumns = inputStackedColumns.flatten()
    inputStackedColumns = np.reshape(inputStackedColumns, (d1_input, d2_input, (h_input-h_kernel+1)//stride * ((w_input-w_kernel+1)//stride), h_kernel*w_kernel ))
    kernel_map_edited = kernel_map.reshape(d_kernel, h_kernel*w_kernel).transpose()
    im2col_conv = np.einsum("ijkl,lm->jkm", inputStackedColumns, kernel_map_edited)
    im2col_conv = im2col_conv.swapaxes(0,2).swapaxes(1,2)
    im2col_conv = im2col_conv.reshape(im2col_conv.shape[0], im2col_conv.shape[1], out_shape[2], out_shape[3])
    return im2col_conv

def convolution(inputMatrix, kernel_map, stride):
    #Vor.: inputMatrix ist vom typ np.darray und hat 4 Dimensionen. kernel_map ist vom Typ np.darray und
    #      hat 3 Dimensionen.
    #Erg.: Bevor -Convolution_strided_img2col- aufgerufen wird, wird -inputMatrix- entsprechend der Funktion getPadding()
    #      gepadded. Danach wird Convolution_strided_img2col() mit der gepaddeten inputMatrix aufgerufen. Das Ergebnis ist
    #      als np.darray geliefert
    
    d1_input, d2_input, h_input, w_input = inputMatrix.shape
    d_kernel, h_kernel, w_kernel = kernel_map.shape

    padding = getPadding(h_kernel, w_kernel, h_input, w_input, stride)
    inputMatrix = np.pad(inputMatrix, ((0,0),(0,0),(0,padding[0]),(0,padding[1])))
    
    return Convolution_strided_img2col(inputMatrix, kernel_map, stride)


def Maxpooling(inputMatrix, kernel_shape, stride):
    #Vor.: inputMatrix ist vom typ np.darray und hat 4 Dimensionen. kernel_shape ist ein zweielementiger Tuple.
    #Erg.: Dimension 3 der Eingabematrix entspricht allen Trainingsbeispielen. Dimension 4 entspricht allen Farbkanälen
    #      für die jeweiligen Trainingsbeispiele. Dimensionen 1 und 2 entsprechen den jeweiligen Pixelwerten der Farbkanäle.
    #      Auf jeden Farbkanal für jedes Trainingsbeispiel wird nun Maxpooling mit einem Kernel der Höhe kernel_shape[0]
    #      und Breite kernel_shape[1] angewandt. Das Ergebnis ist als 4-dimensionaligen np.array geliefert.

    d_kernel_map, d_input, h_input, w_input = inputMatrix.shape
    h_kernel, w_kernel = kernel_shape

    windows = getinputStackedColumns(inputMatrix, kernel_shape, stride)
    
    out_shape = ( d_kernel_map, d_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride, h_kernel, w_kernel)

    maxs = np.max(windows, axis=(4,5))
    maxs = maxs.reshape(d_kernel_map, d_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride)
    return maxs

def Averagepooling(inputMatrix, kernel_shape, stride):
    #Vor.: inputMatrix ist vom typ np.darray und hat 4 Dimensionen. kernel_shape ist ein zweielementiger Tuple.
    #Erg.: Dimension 3 der Eingabematrix entspricht allen Trainingsbeispielen. Dimension 4 entspricht allen Farbkanälen
    #      für die jeweiligen Trainingsbeispiele. Dimensionen 1 und 2 entsprechen den jeweiligen Pixelwerten der Farbkanäle.
    #      Auf jeden Farbkanal für jedes Trainingsbeispiel wird nun Durchschnittspooling mit einem Kernel der Höhe kernel_shape[0]
    #      und Breite kernel_shape[1] angewandt. Das Ergebnis ist als 4-dimensionaligen np.array geliefert.
        
    d_kernel_map, d_input, h_input, w_input = inputMatrix.shape
    h_kernel, w_kernel = kernel_shape
    
    windows = getinputStackedColumns(inputMatrix, kernel_shape, stride)
    
    out_shape = ( d_kernel_map, d_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride, h_kernel, w_kernel)

    means = np.mean(windows, axis=(4,5))
    means = means.reshape(d_kernel_map, d_input, (h_input-h_kernel+1)//stride, (w_input-w_kernel+1)//stride)
    return means

def pooling(inputMatrix, kernel_shape, stride, method):
    #Vor.: inputMatrix ist vom typ np.darray und hat 4 Dimensionen. kernel_shape ist ein zweielementiger Tuple
    #      hat 3 Dimensionen. method ist vom Typ string und hält entweder den Wert 'max' oder 'average' oder andere Wete.
    #      Damit wird die Poolingmethode angegeben, die angewandt werden soll.
    #Erg.: Bevor -Maxpooling- oder -Averagepooling- aufgerufen wird, wird -inputMatrix- entsprechend der Funktion getPadding()
    #      gepadded. Danach wird -Maxpooling- oder -Averagepooling- mit der gepaddeten inputMatrix aufgerufen. Das Ergebnis ist
    #      als np.darray geliefert
    
    d1_input, d2_input, h_input, w_input = inputMatrix.shape
    h_kernel, w_kernel = kernel_shape

    padding = getPadding(h_kernel, w_kernel, h_input, w_input, stride)
    inputMatrix = np.pad(inputMatrix, ((0,0),(0,0),(0,padding[0]),(0,padding[1])))
    
    if method=="average":
        return Averagepooling(inputMatrix, kernel_shape, stride)
    
    elif method=="max":
        return Maxpooling(inputMatrix, kernel_shape, stride)

    else: #Maxpooling ist Standart falls eine andere Methode als Max oder Average angegeben wird.
        return Maxpooling(inputMatrix, kernel_shape, stride)        
```

```python
import numpy as np
import random as rd
from matplotlib import pyplot as plt
from matplotlib import image
import cv2
import csv
import os
import copy

class Model:
    
    imgShape = None
    convout1 = None 
    pooling1 = None
    relu1 = None
    convout2 = None 
    pooling2 = None
    relu2 = None
    x = None
    a2 = None
    a3 = None    
    z2 = None
    z3 = None
    
    w1 = None
    w2 = None 
    kernelmap1 = None
    kernelmap2 = None
    pooling1kernelshape = None
    pooling2kernelshape = None
    
    measurementData_Loss = []
    measurementData_Accuracy = []
    
    def __init__(self, inputArrayShape, n_a2=50, n_a3=2, d_kernel1=5, w_h_kernel1=5, d_kernel2=3, w_h_kernel2=3, poolingmethod="max", w_h_pooling1=2, w_h_pooling2=2, stride_conv1=1, stride_conv2=1, stride_pooling1=1, stride_pooling2=1, learningrate=0.1, batchsize=27):
        #Vor.: Alle Parameter außer poolingmethod und inputArrayShape sind vom Typ int. -inputArrayShape- ist ein 
        #      zweidimensionaliger Tuple. -poolingmethod- ist ein string.
        #Erg.: alle übergebenen Parameter für __inint_- werden als Attribute von der Klasse -Model- initialisiert.
        
        self.imgShape = copy.copy(inputArrayShape)
        self.poolingmethod = poolingmethod
        self.stride_conv1 = stride_conv1 
        self.stride_conv2 = stride_conv2 
        self.stride_pooling1 = stride_pooling1
        self.stride_pooling2 = stride_pooling2
        self.learningrate = learningrate
        self.batchsize = batchsize
        
        self.pooling1kernelshape = (w_h_pooling1,w_h_pooling1)
        self.pooling2kernelshape = (w_h_pooling2,w_h_pooling2)
        
        self.kernelmap1 = np.random.uniform(-1,1,(d_kernel1,w_h_kernel1,w_h_kernel1))
        self.kernelmap2 = np.random.uniform(-1,1,(d_kernel2,w_h_kernel2,w_h_kernel2))
        
        inputArrayShape.insert(0, 1)
        inputArrayShape.insert(0, 1) #input matrix: !one! example with depth !one!
        n_x = self.getNumberOutputNeuronsConvolutionalLayer(inputArrayShape)
        
        self.w1 = np.random.uniform(-1,1,(n_a2, n_x+1))
        self.w2 = np.random.uniform(-1,1,(n_a3, n_a2+1))
        
    def getNumberOutputNeuronsConvolutionalLayer(self, inputArrayShape):
        #Vor.: -inputArrayShape- ist ein zweidimensionaliger Tuple. 
        #Erg.: Testweise würd einmal für ein schwarzes Bild mit der Breite inputArrayShape[1] und der Höhe inputArrayShape[0]
        #      durch die zwei faltenden Schichten geführt. Im Folgenden werden die Anzahl an Pixel der resultierenden 
        #      Matrix nach den 2 falenden Schichten bestimmt. Dies entspricht der Anzahl an Eingangsneuronen der 2 
        #      regulären neuronalen Schichten. Die Anzahl an benötigten Eingabeneuronen bzw. Somata für die regulären 
        #      neuronalen Schichten ist als int geliefert.
        
        self.input = np.zeros((inputArrayShape[0], inputArrayShape[1], inputArrayShape[2], inputArrayShape[3]))
        self.__convolutional_layers(self.input)
        return (self.x.shape[0])
        
    def __convolutional_layers(self, input):
        #Vor.: -input- ist ein np.darray mit 4 Dimensionen.
        #Eff.: Die Reizweiterleitung innerhalb der 2 faltenden Schichten wird realisiert. Die Attribut des Datenmodels
        #      (z.B. self.poolingout1) werden entsprechend mit der Reizweiterleitung geändert.
        
        self.convout1 = convolution(self.input, self.kernelmap1, self.stride_conv1) 
        self.reluout1 = RELU_Matrixoperation(self.convout1)
        self.poolingout1 = pooling(self.reluout1, self.pooling1kernelshape, self.stride_pooling1, self.poolingmethod)

        self.convout2 = convolution(self.poolingout1, self.kernelmap2, self.stride_conv2) 
        self.reluout2 = RELU_Matrixoperation(self.convout2)
        self.poolingout2 = pooling(self.reluout2, self.pooling2kernelshape, self.stride_pooling2, self.poolingmethod)
        
        self.x = self.poolingout2.reshape(self.poolingout2.shape[0], self.poolingout2.shape[1], self.poolingout2.shape[2]*self.poolingout2.shape[3])
        self.x = np.concatenate(self.x, axis=1).transpose()
        
    def propagateForward(self, input):
        #Vor.: -input- ist ein np.darray mit 4 Dimensionen. Dimension 3 der Eingabematrix entspricht allen Trainingsbeispielen.
        #      Dimension 4 entspricht allen Farbkanälen für die jeweiligen Trainingsbeispiele. Dimensionen 1 und 2 
        #      entsprechen den jeweiligen Pixelwerten der Farbkanäle.
        #Eff.: Die Reizweiterleitung innerhalb der 2 faltenden und 2 regulären neuronalen Schichten werden für alle Trainingsbeispiele 
        #      in Dimension 4 von -input- parallel realisiert. Hierfür werden alle Spaltenvektoren (Eingabevektoren) für alle Bilder 
        #      horizontal aneinander 'geklebt'. Die Reizweiterleitung erfolgt nun wie im Exposé beschrieben, bloß mit dem Skalarprodukt
        #      aus der Eingabe!matrix! und der Synapsen-Matrix
        #Erg.: die Ausgabematrix ist geliefert. Sie besteht aus den Spaltenvektoren für die Ausgabe aller Spaltenvektoren (Eingabevektoren).
        #      Diese Spaltenvektoren sind wieder horizontal 'aneinandergeklebt'. Jeder Ausgabe-Spaltenvektor besteht aus 2 Soma-Werten:
        #      1. Wert: Wahrscheinlichkeit dafür, dass Patient krank (0 bis 1)
        #      2. Wert: Wahrscheinlichkeit dafür, dass Patient gesund (0 bis 1)
        
        self.input = input.reshape(input.shape[0], 1, input.shape[1], input.shape[2]).swapaxes(0,1)
        self.__convolutional_layers(self.input)
        self.x = np.pad(self.x, ((1,0),(0,0)), constant_values=1)
        
        self.z2 = np.dot(self.w1, self.x)
        self.a2 = sigmoid(self.z2)
        self.a2 = np.pad(self.a2, ((1,0),(0,0)), constant_values=1)
        
        self.z3 = np.dot(self.w2, self.a2)
        self.a3 = softmax(self.z3)
        
        return self.a3
        
        
    def propagateBackward(self, input, y_hat):
        #Vor.: -input- ist ein np.darray mit 4 Dimensionen. y-hat (Wahrheitswerte) hat 2 Dimensionen 
        #Eff.: Die Optimierung der Parameter (v(2), v(1)) wird hier realisiert (nicht sequenziel mit for-Loops wie in 
        #      Präsentation beschrieben, sondern parallel mit Operationen der linearen Algebra). 
        
        self.propagateForward(input)
        pd_L_wrt_a3 = 1/input.shape[0] * (self.a3 - y_hat)
        pd_a3_wrt_z3 = softmax(self.z3)*(1 - softmax(self.z3))
        pd_z3_wrt_w2 = self.a2
        pd_L_wrt_w2 = np.dot(pd_L_wrt_a3 * pd_a3_wrt_z3, pd_z3_wrt_w2.T)
        
        pd_z3_wrt_a2 = self.w2
        pd_a2_wrt_z2 = sigmoid(self.z2)*(1 - sigmoid(self.z2))
        pd_z2_wrt_w1 = self.x
        pd_L_wrt_w1 = np.dot( np.dot((pd_L_wrt_a3 * pd_a3_wrt_z3).T, pd_z3_wrt_a2[:,1:]).T * pd_a2_wrt_z2, pd_z2_wrt_w1.T)

        self.w2 = self.w2 - self.learningrate * pd_L_wrt_w2
        self.w1 = self.w1 - self.learningrate * pd_L_wrt_w1
        
    def getDataBatch(self, path, filenames):
        #Vor.: -path- ist ein String, der auf ein gültiges Verzeichnis verweist, indem sich alle
        #      Trainingsbilder befinden. -filenames- ist eine Liste, die alle Dateinamen der Trainingsbilder beinhaltet.
        #Erg.: Die Wahrheitswerte und Eingabematrizen für jedes Trainingsbeispiel in der Liste -filenames- sind gebündelt
        #      in einem zweidimensionaligen Tuple geliefert.
        
        uniform_size = self.imgShape[::-1]
        
        data_x = []
        data_y = []
        
        for i in filenames:
            
            img = image.imread(path + "//" + i)
            if len(img.shape)==3:
                img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
            resized_img = cv2.resize(img, dsize=uniform_size, interpolation=cv2.INTER_CUBIC)
            data_x.append(resized_img)


            y = None
            if "y" in i:
                y = np.array([1,0])
            else:
                y = np.array([0,1])
            data_y.append(y)
            
        data_x = np.array(data_x).astype(np.float)  
        data_y = np.array(data_y).astype(np.float).transpose()
        
        return(data_x, data_y)
        
    def train(self, epochs, trainpath, testpath):
        #Vor.: -epochs- ist vom Typ int. -testpath- und -trainpath- sind gültige Verzeichnisnamen 
        #      und verweisen auf das Verzeichnis aller Trainings- und Testbilder (MRT-Scans).
        #Eff.: Das neuronale Netz wird mit allen angegebenen Trainingsbildern (MRT-Scans) in -epochs- Trainingsiterationen trainiert.
        
        trainfilenames = os.listdir(trainpath)
        testfilenames = os.listdir(testpath)
        
        np.random.shuffle(trainfilenames)
        np.random.shuffle(testfilenames)
        
        batch_counter = 0
        for i in range(epochs):
            filenamesBatch = trainfilenames[batch_counter:batch_counter+self.batchsize]
            
            data = self.getDataBatch(trainpath, filenamesBatch)
            input = data[0]
            y_hat = data[1]
            self.propagateBackward(input, y_hat)
            loss = 1/input.shape[0] * ((self.a3 - y_hat)**2).sum()
            
            data_test = self.getDataBatch(testpath, testfilenames[:100])
            accuracy = self.test(data_test[0], data_test[1])
            print("Epoch " + str(i) + " done. Loss: " + str(loss) + " Accuracy: " + str(accuracy*100) + " %")
            self.measurementData_Accuracy.append(accuracy)
            self.measurementData_Loss.append(loss)
            
            batch_counter += self.batchsize
            if (batch_counter>=2700-self.batchsize):
                batch_counter = 0
            
    def test(self, input, y_hat):
        #Vor.: -input- ist ein np.darray mit 4 Dimensionen. y-hat (Wahrheitswerte) hat 2 Dimensionen.
        #Erg.: Die Genauigkeit des Models für die Eingabedaten -input- und den entsprechenden Wahrheitswerten
        #      int -y-_hat- wird getestet und als Float-Wert geliefert (0 bis 1).
        
        fx = self.propagateForward(input)
        number_correct = 0
        number_overall = fx[:,:100].shape[1]
        for i in range(number_overall):
            arrayfx = fx[:,i:i+1]
            arrayy = y_hat[:,i:i+1]
           
            indexmaxfx = np.argmax(arrayfx, axis=0)
            indexmaxy = np.argmax(arrayy, axis=0)
            if (indexmaxfx[0]==indexmaxy[0]):
                number_correct += 1
                
        return number_correct/number_overall
    
    def getMeasurementData(self):
        #Vor.: -
        #Erg.: Der Verlauf des mittleren quadrierten Fehlers pro Trainingsiteration (in Form einer Liste) und 
        #      der Verlauf der Genauigkeit pro Trainingsiteration (in Form einer Liste) werden gebündelt als 
        #      Tuple geliefert. Der Verlauf beider Größen kann dann mit dem matplotlib-Modul visualisiert werden.
        
        return (self.measurementData_Loss, self.measurementData_Accuracy)
    
    def saveweights(self):
        #Vor.: -
        #Eff.: Die Matrix aller Synapsen und Kernels ist in einer externen Datei im aktuellen Verzeichnis gespeichert.
        
        np.save("kernelmap1.npy", self.kernelmap1)
        np.save("kernelmap2.npy", self.kernelmap2)
        np.save("w1.npy", self.w1)
        np.save("w2.npy", self.w2)

    def loadweights(self):
        #Vor.: -
        #Eff.: Die Matrix aller Synapsen und Kernels, die als externe Datei gespeichert sind, werden als
        #      aktuelle Synapsen und Kernels für das Model übernommen. 
        #      So können die Parameter eines Models, die bereits optimiert sind, abgespeichert werden und später
        #      wieder ausgelesen werden. So kann ein bereits trainiertes Model nach Schließen der ipython-Umgebung
        #      wieder hergestellt werden.
        
        try:
            self.kernelmap1 = np.load("kernelmap1.npy")
            self.kernelmap2 = np.load("kernelmap2.npy")
            self.w1 = np.load("w1.npy")
            self.w2 = np.load("w2.npy")

        except Exception as e:
            print(e)
```

```python
from random import *

tracking_data_models = []
tracking_data_accuracy = []
tracking_data_loss = []

while True:
    n_a2 = randint(1,15)*1000
    n_a3 = 2
    d_kernel1 = randint(1,6)
    w_h_kernel1 = randint(1,6)
    d_kernel2 = randint(1,6)
    w_h_kernel2 = randint(1,6)
    poolingmethod = randint(1,2) #1--> max, 2--> average
    w_h_pooling1 = randint(1,6)
    w_h_pooling2 = randint(1,6)
    stride_conv1 = randint(1, w_h_kernel1)
    stride_conv2 = randint(1, w_h_kernel2)
    stride_pooling1 = randint(1, w_h_pooling1)
    stride_pooling2 = randint(1, w_h_pooling2)
    learningrate = randint(1,10)*0.1
    batchsize = 27
    
    try:
        
        if poolingmethod==1:
            poolingmethod="average"
        else:
            poolingmethod="max"
        
        model_list = [n_a2, n_a3, d_kernel1, w_h_kernel1, d_kernel2, w_h_kernel2, poolingmethod, w_h_pooling1, w_h_pooling2, stride_conv1, stride_conv2, stride_pooling1, stride_pooling2, learningrate, batchsize]
        
        print("Model ", model_list, " started.")
                
        m = Model([175, 167], n_a2, n_a3, d_kernel1, w_h_kernel1, d_kernel2, w_h_kernel2, poolingmethod, w_h_pooling1, w_h_pooling2, stride_conv1, stride_conv2, stride_pooling1, stride_pooling2, learningrate, batchsize)
        m.train(200, "C://Users//Benni//Desktop//PK//Dateset//yes_and_no//train", "C://Users//Benni//Desktop//PK//Dateset//yes_and_no//test")
        
        accuracy_list = m.measurementData_Accuracy
        loss_list = m.measurementData_Loss
        tracking_data_models.append(model_list)
        tracking_data_accuracy.append(accuracy_list)
        tracking_data_loss.append(loss_list)
        
        print("Model ", model_list, " done. [", len(tracking_data_models), "] Max: ", max(accuracy_list), "\n")
        
    except Exception as e:
        print(e)
        continue
```
